## Wprowadzenie, notacja, definicje <a class="anchor" id="big0"></a>

Według internetowej wersji słownika języka polskiego[$^{1}$](#bib_1) znaczenie słowa klaster jest na tyle różnorodne, że obejmuje wiele dziedzin życia codziennego. W definicji zwraca się uwagę na istnienie więcej niż jednego atomu, dźwięku bądź firmy, a także na podobieństwo pomiędzy tworzonymi przez nie grup zwanych klastrami.

Klasteryzacja (analiza skupień, grupowanie) to dziedzina zainteresowań dziedziny z pogranicza matematyki i statystyki - uczenia maszynowego. W najbardziej ogólnym ujęciu klasteryzację można zaliczyć do działu uczenia nienadzorowanego (ang. *unsupervised learning*), czyli takiego gdzie algorytm sam musi zdecydować o przynależności elementu do grupy.[$^{2}$](#bib_3a)  

W ujęciu matematycznym problem dotyczy $N$ punktów $X$ = ${x_1, x_2, ..., x_N}$ pochodzących z $D$-wymiarowej przestrzeni euklidesowej. Zazwyczaj używa się odległości euklidesowej, ale założenia analizy nie wykluczają przypadku danych kategorycznych, dla których używa się odległości Hamminga.[$^3$](#bib_8) Celem analizy jest znalezienie, abstrahując od tego czy wartość jest znana a priori, $K$ grup (klastrów). O pojedynczym grupowaniu można myśleć jak o zbiorze punktów, które są blisko siebie nawzajem oraz jak najdalej od innych skupisk punktów. Dokładniej przedstawiona idea klastra zasada się na zdefiniowaniu wektorów $\mu_k$, gdzie $k = 1, 2, ..., K$. Każdy z tych wektorów stanowi prototyp (wartość reprezentacyjną) swojego klastra. Optymalny wynik algorytmu grupowania otrzymuje się w wyniku zminimalizowania odległości pomiędzy punktami ze zbioru $X$ a najbliższymi im wektorami $\mu_k$.[$^4$](#bib_3b)

Formalny sposób kodowania przynależności punktu do klastra wymaga wprowadzenia zmiennej $r_{nk}$ $\epsilon$ $\{0, 1\}$. W przypadku, gdy $n$-ty punkt należy do $k$-tego klastra piszemy $r_{nk} = 1$ oraz $r_{nj} = 0$ dla $j\neq k$. Wspomniana wcześniej funkcja do minimalizacji, której dążymy to:

<br><center>$\displaystyle J = \sum_{n=1}^{N}\sum_{k=1}^{K} r_{nk} ||{x_n - \mu_k}||^2 $</center>

Sądząc po definicji słowa "klaster" można oczekiwać, że algorytmy grupowania danych znajdą zastosowanie w wielu dziedzinach nauki. Następujące przykłady wskazują na szerokie zastosowanie idei opisanej w powyższym akapicie. Analiza skupień używana jest w odkrywaniu ciał niebieskich[$^5$](#bib_5), segmentacji klientów w celu dopasowania reklam im przesyłanych[$^6$](#bib_6)$^{,}$[$^7$](#bib_10), badaniach biologicznych dotyczących różnych rodzajów komórek znajdujących w organizmach żywych[$^8$](#bib_7), dopasowaniu działań pracowników IT do zgłaszanych usterek[$^9$](#bib_9), a także zwalczaniu przestępczości.[$^{10}$](#bib_11). 

Opisy sztandarowej metody klastrowania, która opiera się na doborze $k$ punktów (niezależnie czy są to średnie czy mediany reprezentujące swoje klastry), zwracają uwagę na problem doboru liczby skupień oraz początkowe rozmieszczenie tych punktów w badanej przestrzeni.[$^{11}$](#bib_12)$^,$[$^{12}$](#bib_4a) W kontraście do metod opartych na wyborze punktów reprezentacyjnych (średnie, mediany) stoją te bazujące na hierarchicznym podejściu do zjawiska odmienności (ang. *dissimilarity*) wśród badanych punktów.[$^{13}$](#bib_4b) Hierarchiczne klastrowanie zakłada istnienie dwóch przypadków granicznych: jeden z nich dotyczy sytuacji, gdy wszystkie punkty znajdują się w jednym klastrze, natomiast drugi takiej, kiedy każdy punkt stanowi swój własny klaster. 

Jedno duże grupowanie to sytuacja wyjściowa dla algorytmów podziałowych (ang. *divisive*). W każdej iteracji grupy dzielone są na dwie części, tak aby dystans pomiędzy stworzynymi grupami był maksymalny. Algorytm kończy działanie, gdy dane zostaną podzielone na $n$ grupowań, gdzie $n$ to liczba obserwacji.

Gdyby przyjąć strategię wychodzącą z $n$ klastrów mamy do czynienia z podejściem aglomeratywnym (ang. *agglomerative*). Z każdym powtórzeniem algorytmu dwa najbardziej do siebie podobne klastry (w pierwszej iteracji pojedyncze punkty) są ze sobą łączone w pary. Program kończy bieg, gdy uzyskuje jedno grupowanie obejmujące wszystkie punkty.

Poniższy przykład pokazuje zastosowanie algorytmu hierarchicznego na danych dotyczących przestępczości w poszczególnych stanach Stanów Zjednoczonych.[$^{14}$](#bib_17)$^{,}$[$^{15}$](#bib_18) Liczba klastrów zależy od punktu, w którym osoba analizująca dane odetnie drzewo (punkt startowy drzewa znajduje się na górze dendrogramu) - w poniżej prezentowanym przypadku wybrano wysokość odcięcia gwarantującą wskazanie przez algorytm czterech klastrów.

Dokładniejsza analiza problemu grupowania aglomeratywnego skłania do przemyśleń na temat sposobu w jaki można badać odmienność punktów w dwóch różnych klastrach. Załóżmy, że $G$ i $H$ to dwie różne grupy punktów przestrzeni, którą badamy i której elementy próbujemy zaklasyfikować do klastrów. Istnieją trzy sposoby określenia miary odmienności.[$^{16}$](#bib_4c) Korzystając z *single linkage* bierzemy pod uwagę dwa punkty z grup $G$ i $H$ (po jednym z każdej) i jako odmienność określamy odległość $d_{ii'}$ pomiędzy dwoma najbliższymi punktami:

<br><center>$ \displaystyle d_{SL}(G,H) = \min_{i \epsilon G, i' \epsilon H}  d_{ii'}$</center>

*Complete linkage* bierze pod uwagę dwa najbardziej oddalone od siebie punkty:

<br><center>$ \displaystyle d_{CL}(G,H) = \max_{i \epsilon G, i' \epsilon H}  d_{ii'}$</center>

W podejściu *group average* bierze się pod uwagę wszystkie odległości pomiędzy punktami z obu grup:

<br><center>$\displaystyle d_{GA}(G,H) = \frac{1}{N_G N_H} \sum_{i \epsilon G} \sum_{i' \epsilon H} d_{ii'}$, gdzie $N_G$ oraz $N_H$ to liczebności odpowiednio grupy G oraz grupy H. </center>

Zakładając, że dane, które chcemy podzielić na klastry, mają postać mocno skupionych chmur znajdujących się od siebie nawzajem we względnie (w stosunku do średnich chmur) dużej odległości wszystkie trzy powyższe metody powinny doprowadzić do takich samych wyników grupowania. W przypadku metody *single linkage* pojawia się problem rozciągniętych w przestrzeni punktów. Nawet, gdy wewnętrzna średnica chmury punktów zdefiniowana jako <br><br><center>$\displaystyle D_G = \max_{i \epsilon G, i' \epsilon G} d_{ii'}$</center><br> jest stosunkowo duża do uznania dwóch grup za podobne wystarczy, aby dwa skrajne punkty były blisko siebie. Właśnie tą metodę porównywania odmienności grup wykorzystuje wariant drzew o minimalnej rozpiętości (ang. *minimum spanning tree*) użyty w poniższych rozważaniach dotyczących analizy skupień.[$^{17}$](#bib_13)

Zgodnie z filozofią grupowania hierarchicznego można analizować problem klasteryzacji w kategoriach zbioru wierzchołków, które mają zostać połączone tak, aby tworzyły drzewo. Analiza skupień bazuje na podobieństwie elementów wewnątrz klastra oraz odmienności poszczególnych klastrów od siebie.  

Drzewo o minimalnej rozpiętości łączy wszystkie wierzchołki w taki sposób, że całkowity koszt (w naszym przypadku suma odległości między połączonymi punktami) jest minimalny. Jeżeli zbiór punktów jest klastrowalny wystarczy wprowadzić pewną długość, powyżej której połączenia pomiędzy punktami odcinamy. W ten sposób powstają grupowania oddzielone od siebie w wyraźny sposób.

### Sztuczne dane bez hałasu <a class="anchor" id="medium1"></a>

```python
no_centers = 4
no_points = 200

Markdown("""
Poni&#380;szy przyk&#322;ad, w kt&oacute;rym korzystamy z (pseudo)losowo wygenerowanych skupie&#324; punkt&oacute;w, prezentuje zagadnienie konstrukcji drzewa o minimalnej rozpi&#281;to&#347;ci dla najprostszego przypadku, czyli wyra&#378;nie oddzielnych chmur punkt&oacute;w. 
Analizowane s&#261; cztery losowo wygenerowane skupienia - ka&#380;de po {sing_clus} punkt&oacute;w. Wizualna inspekcja prowadzi do wniosku, &#380;e nale&#380;&#261; do czterech r&oacute;&#380;nych klastr&oacute;w:
""".format(sing_clus = no_points/no_centers))
```

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
X, y = make_blobs(no_points, centers=no_centers, random_state=8)
plt.scatter(X[:, 0], X[:, 1], c='lightblue')
```

```python
import numpy as np
Markdown("""Algorytm MST poprawnie wskaza&#322; {no_clus} klastry, ka&#380;dy oznaczony innym kolorem:""".format(no_clus = len(np.unique(y))))
```

```python
!pip install mst_clustering
```

```python
from mst_clustering import MSTClustering
model = MSTClustering(cutoff_scale=1.6, approximate=False)
labels = model.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow')
```

Poniższe wykresy prezentują ideę działania klasteryzacji za pomocą drzewa o minimalnej rozpiętości:

```python
def plot_mst(model, cmap='rainbow'):
    X = model.X_fit_
    fig, ax = plt.subplots(1, 2, figsize=(16, 6), sharex=True, sharey=True)
    for axi, full_graph, colors in zip(ax, [True, False], ['lightblue', model.labels_]):
        segments = model.get_graph_segments(full_graph=full_graph)
        axi.plot(segments[0], segments[1], '-k', zorder=1, lw=1)
        axi.scatter(X[:, 0], X[:, 1], c=colors, cmap=cmap, zorder=2)
        axi.axis('tight')
    
    ax[0].set_title(u'Pełne drzewo o minimalnej rozpiętości', size=16)
    ax[1].set_title(u'"Przycięte" drzewo o minimalnej rozpiętości', size=16)
    
plot_mst(model)
```

Przy tworzeniu algorytmów analizy skupień warto mieć na uwadze zjawisko "hałasu", które, gdy zignorowane, prowadzi do ukrycia rzeczywistych wzorców występujących w oczyszczonych z zanieczyszczeń danych. Problem ten jest szczególnie istotny, gdy chodzi o dane empiryczne. Podczas ich zbierania istnieje wiele okazji do tego, żeby czy to osoba prowadzącą badanie, czy administrator bazy danych czy programista wprowadził niechcący błędne dane.[$^{18}$](#bib_14)

### Sztuczne dane z hałasem <a class="anchor" id="medium2"></a>

Poniższy przykład ukazuje zastosowanie MST na (pseudo)losowo wygenerowanym zbiorze danych zawierającym hałas:

```python
rng = np.random.RandomState(int(100 * y[-1]))
noise = -14 + 28 * rng.rand(200, 2)

X_noisy = np.vstack([X, noise])
y_noisy = np.concatenate([y, np.full(200, -1, dtype=int)])

plt.scatter(X_noisy[:, 0], X_noisy[:, 1], c='lightblue', cmap='spectral_r')
plt.xlim(-15, 15)
plt.ylim(-15, 15);
```

```python
noisy_model = MSTClustering(cutoff_scale=1)
noisy_model.fit(X_noisy)
cluster_number = len(np.unique(noisy_model.labels_))

Markdown("""
Pierwsze wyniki dzia&#322;ania algorytmu na danych z ha&#322;asem nie s&#261; obiecuj&#261;ce. Minimalne drzewo zosta&#322;o przyci&#281;te w 
taki spos&oacute;b, &#380;e rozpozna&#322;o {clus_num} klastr&oacute;w mimo, &#380;e dane zawieraj&#261; jedynie cztery grupowania otoczone ha&#322;asem:
""".format(clus_num=cluster_number))
```

```python
plot_mst(noisy_model)
```

```python
min_size = 25

Markdown("""
Podejmuj&#261;c si&#281; analizy skupie&#324; w danych z ha&#322;asem przy u&#380;yciu drzewa o minimalnej rozpi&#281;to&#347;ci warto zastrzec minimalny rozmiar klastra. Ta metoda pozwala na wyeliminowanie klastr&oacute;w wynikaj&#261;cych z losowego u&#322;o&#380;enia niezwi&#261;zanych ze sob&#261; punkt&oacute;w obok siebie. W poni&#380;szym przyk&#322;adzie minimalny rozmiar zosta&#322; ustalony na {min_size}:
""".format(min_size=min_size))
```

```python
noisy_model_clus_size = MSTClustering(cutoff_scale=1, min_cluster_size=min_size)
noisy_model_clus_size.fit(X_noisy)
plot_mst(noisy_model_clus_size, cmap='CMRmap')
```

## Dane rzeczywiste: irysy <a class="anchor" id="big2"></a>

Zestawem danych, na którym sprawdzimy skuteczność grupowania za pomocą drzewa o minimalnej rozpiętości jest zestaw pomiarów trzech gatunków irisów (Iris setosa, Iris virginica oraz Iris versicolor).[$^{19}$](#bib_21) Dane zawierają wyrażone w centymetrach pomiary długości oraz szerokości działek kielicha oraz płatków. Każdy gatunek jest reprezentowany przez 50 sztuk kwiatów.

Dane zawierają czterowymiarową przestrzeń zmiennych niezależnych. Jako, że gatunki, do których należą poszczególne obserwacje, są znane *a priori*, po wykonaniu analizy metodą drzewa o minimalnej rozpiętości będzie można dokonać pomiaru dokładności jego działania. W celu graficznego przedstawienia klasyfikacji poszczególnych obserwacji należy posłużyć się jedną z technik redukcji wymiarów. 

Rozsądnym, w przypadku wymiarów roślin, jest założyć, że wymiary pojedynczej rośliny są ze sobą skorelowane.[$^{20}$](#bib_16a) Empirycznym dowodem może być w tym przypadku macierz korelacji poszczególncych cech:

```python
from sklearn import datasets
from sklearn.decomposition import PCA
import pandas as pd
```

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(data = iris.data,
                      index = range(0,len(iris.data)),
                      columns = iris.feature_names)

iris_df["target"] = iris.target

iris_df.target = np.where(iris_df.target == 0, "setosa", np.where(iris_df.target == 1, "versicolor", "virginica"))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

iris_corr = iris_df.iloc[:, 0:4].corr()

masking = np.zeros_like(iris_corr, dtype=np.bool)
masking[np.triu_indices_from(masking)] = True

f, ax = plt.subplots(figsize=(11, 9))

cmap = sns.diverging_palette(220, 10, as_cmap=True)

sns.heatmap(iris_corr, mask=masking, cmap=cmap, vmin=-1, vmax=1, center=0,
            square=True, linewidths=.75, cbar_kws={"shrink": .75})
```

Analiza głównych składowych pozwoli zamienić czterowymiarową przestrzeń na przestrzeń maksymalnie trzywymiarową, znajdując pewne zmienne nieskorelowane, które są liniowymi kombinacjami pierwotnych czterech zmiennych.

```python
pca = PCA(n_components=3)
pca.fit(iris.data)
pca.explained_variance_ratio_


Markdown("""
Dwie pierwsze g&#322;&oacute;wne sk&#322;adowe, tworz&#261;ce now&#261; przestrze&#324;, w kt&oacute;rej zostan&#261; 
osadzone pierwotne pomiary iris&oacute;w, t&#322;umacz&#261; a&#380; {tot_var_2} zmienno&#347;ci zawartej w oryginalnym zbiorze danych: {pca_1}, {pca_2} oraz {pca_3} 
odpowiednio pierwsza, druga oraz trzecia g&#322;&oacute;wna sk&#322;adowa.
""".format(tot_var_2 = '{0:.2%}'.format(sum(pca.explained_variance_ratio_[:3])), 
           pca_1 = '{0:.2%}'.format(pca.explained_variance_ratio_[0]), 
           pca_2 = '{0:.2%}'.format(pca.explained_variance_ratio_[1]),
           pca_3 = '{0:.2%}'.format(pca.explained_variance_ratio_[2])))
```

Obserwacje osadzone w nowej przestrzeni zdefiniowanej przez trzy główne składowe skłaniają do wniosku, że poszczególne gatunki irisów są separowalne w oparciu o metodę drzewa o minimalnej rozpiętości, ponieważ stanowią zbite chmury punktów. Problematyczne może okazać się oddzielenie od siebie grup versicolor oraz virginica, ponieważ jak widać na załączonym interaktywnym wykresie, grupy te są bardzo blisko siebie:

```python
import plotly
import plotly.graph_objs as go

pca_iris = pca.transform(iris.data) #zawiera 150 punktów określonych przez współrzędne w PCA

setosa = go.Scatter3d(
    x=[row[0] for row in pca_iris[:50]],
    y=[row[1] for row in pca_iris[:50]],
    z=[row[2] for row in pca_iris[:50]],
    mode='markers',
    marker=dict(
        size=12,
        line=dict(
            color='rgba(217, 217, 217, 0.14)',
            width=0.5
        ),
        opacity=0.8
    ),
    name = "setosa"
)

versicolor = go.Scatter3d(
    x=[row[0] for row in pca_iris[50:100]],
    y=[row[1] for row in pca_iris[50:100]],
    z=[row[2] for row in pca_iris[50:100]],
    mode='markers',
    marker=dict(
        size=12,
        line=dict(
            color='rgba(10, 25, 27, 0.14)',
            width=0.5
        ),
        opacity=0.8
    ),
    name = "versicolor"
)

virginica = go.Scatter3d(
    x=[row[0] for row in pca_iris[100:150]],
    y=[row[1] for row in pca_iris[100:150]],
    z=[row[2] for row in pca_iris[100:150]],
    mode='markers',
    marker=dict(
        size=12,
        line=dict(
            color='rgba(100, 21, 250, 0.14)',
            width=0.5
        ),
        opacity=0.8
    ),
    name = "virginica"
)

data = [setosa, versicolor, virginica]

layout = go.Layout(
    scene = dict(
                    xaxis = dict(
                        title='Pierwsza główna składowa'),
                    yaxis = dict(
                        title='Druga główna składowa'),
                    zaxis = dict(
                        title='Trzecia główna składowa'),),
    
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0
    ),
    legend=dict(orientation="h")
)

fig = go.Figure(data=data, layout=layout)

plotly.offline.init_notebook_mode(connected=True)

plotly.offline.iplot(fig, filename='simple-3d-scatter')
```

Zastosujmy algorytm drzewa o minimalnej rozpiętości do problemu analizy skupień wsród irysów. Krokiem, który umożliwi sprawdzenie skuteczności grupowania przez algorytm MST jest zestawienie zliczeń poszczególnych metek (ang. *labels*, czyli oznaczeń pojedyncznych pomiarów kwiatów) w tablicy pomyłek (ang. *confusion matrix*, która jest kwadratową macierzą o wymiarach $kxk$, gdzie to $k$ to liczba metek). Zerowy błąd klasyfikacji uzyskujemy, gdy wartości większe od zera znajdują się jedynie na przekątnej macierzy. 

W poniższym przypadku całościowy sukces algorytm odniósł jedynie w przypadku gatunku setosa, którego skupisko jak wcześniej zauważono, jest oddalone od pozostałych dwóch skupisk. Drzewo o minimalnej rozpiętości niefortunnie połączyło obserwacje dwóch pozostałych gatunków tak, żeby zostały uznane za pojedynczy klaster: 

```python
from sklearn.metrics import confusion_matrix
model_iris = MSTClustering(cutoff_scale = 1.5)
labels_iris = model.fit_predict(iris.data)
con_matrix = confusion_matrix(iris.target, labels_iris)
```

```python
%matplotlib inline

import itertools

def plot_confusion_matrix(cm, classes,
                          title='',
                          cmap=plt.cm.Blues):

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], 'd'),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    #plt.tight_layout()
    plt.ylabel('Rzeczywisty gatunek')
    plt.xlabel('Modelowany gatunek')
    plt.grid(False)
```

```python
plot_confusion_matrix(con_matrix, classes = iris.target_names)
```

## Wnioski <a class="anchor" id="big3"></a>

Na zakończenie, należy zauważyć, że analizowany powyżej algorytm analizy skupień oparty o konstrukcję drzewa o minimalnej rozpiętości sprawdza się w warunkach danych, tworzących wyraźnie skupiska. Na uwagę zasługuje fakt, że nie potrzebna jest żadna wcześniejsza wiedza osoby analizującej problem na temat przestrzeni, z której dane pochodzą. Skuteczność algorytmu w modelowaniu przestrzeni z hałasem wymaga jednak pewnej wcześniejszej wiedzy o minimalnym rozmiarze klastra. Idea odcinania krawędzi o wadze (długości) przekraczającej oczekiwania analityka, co do odłegłości pomiędzy punktami wewnątrz danego klastra, nie sprawdza się jednak w warunkach, gdy dwa lub więcej skupisk (o których wiemy *a priori*, że istnieją) jest zbyt blisko siebie jak w przykładzie z klasteryzacją gatunków irysów.
